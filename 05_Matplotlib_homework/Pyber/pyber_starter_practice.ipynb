```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df =pd.read_csv(city_data_to_load)
ride_df =pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_df = pd.merge(ride_df, city_df, on = "city")

# Display the data table for preview
merge_df
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_type = merge_df.loc[(merge_df["type"] == "Urban")]
urban_grouped_df = urban_type.groupby(["city"])
urban_driver_count = urban_grouped_df["driver_count"].unique()
urban_number_rides = urban_grouped_df["ride_id"].count()
urban_average_fare = urban_grouped_df["fare"].mean()
int_urban_driver_count = urban_driver_count.astype(int)
```

```python
suburban_type = merge_df.loc[(merge_df["type"] == "Suburban")]
suburban_grouped_df = suburban_type.groupby(["city"])
suburban_driver_count = suburban_grouped_df["driver_count"].unique()
suburban_number_rides = suburban_grouped_df["ride_id"].count()
suburban_average_fare = suburban_grouped_df["fare"].mean()
int_suburban_driver_count = suburban_driver_count.astype(int)
```

```python
rural_type = merge_df.loc[(merge_df["type"] == "Rural")]
rural_grouped_df = rural_type.groupby(["city"])
rural_driver_count = rural_grouped_df["driver_count"].unique()
rural_number_rides = rural_grouped_df["ride_id"].count()
rural_average_fare = rural_grouped_df["fare"].mean()
int_rural_driver_count = rural_driver_count.astype(int)
```

```python
plt.scatter(rural_number_rides,rural_average_fare, s = int_rural_driver_count*10,c= "yellow" , edgecolors='black', alpha=0.4, label = "Rural")
plt.scatter(suburban_number_rides,suburban_average_fare, s = int_suburban_driver_count*10,c= "skyblue" , edgecolors='black', alpha=0.4, label = "Suburban")
plt.scatter(urban_number_rides,urban_average_fare, s = int_urban_driver_count*10, c= "red" , edgecolors='black', alpha=0.4, label = "Urban")



plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Numbers of Rides (Per City))")
plt.ylabel("Average Fare($)")

lgnd = plt.legend(loc='upper right',prop={'size': 10},title = "city type")
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]

plt.text(43,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

plt.grid()
```

```python
plt.savefig("Output/Pyber.png")
plt.show()
```

## Total Fares by City Type

```python
total_fares = merge_df.groupby(["type"])["fare"].sum()
labels = ["Rural","Suburban","Urban"]
colors = ["gold","lightskyblue","lightcoral"]
explode = [0,0, 0.1]
plt.title("% of Total Fares By City Types")
plt.pie(total_fares, explode=explode, labels=labels, colors=colors, autopct="%.1f%%",shadow=True, startangle=160)
plt.savefig("Output/percets_total_fares_.png")
plt.show()
```

## Total Rides by City Type

```python
total_rides = merge_df.groupby(['type'])['ride_id'].count()
labels = ["Rural","Suburban","Urban"]
colors = ["gold","lightskyblue","lightcoral"]
explode = [0,0, 0.1]
plt.title("% of Total Rides By City Types")
plt.pie(total_rides, explode=explode, labels=labels, colors=colors, autopct="%.1f%%",shadow=True, startangle=160)
plt.savefig("Output/percets_total_rides.png")
plt.show()
```

## Total Drivers by City Type

```python
total_drivers = city_df.groupby(['type'])['driver_count'].sum()
labels = ["Rural","Suburban","Urban"]
colors = ["gold","lightskyblue","lightcoral"]
explode = [0,0, 0.1]
plt.title("% of Total Drivers By City Types")
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors, autopct="%.1f%%",shadow=True, startangle=160)
plt.savefig("Output/percets_total_drivers.png")
plt.show()
```
